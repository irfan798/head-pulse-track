### Butterworth filter

```python
from scipy import interpolate, signal, optimize
from scipy.fftpack import fft, ifft, fftfreq, fftshift
import numpy as np
import matplotlib.pyplot as plt
```

```python
fs = 30  # Sampling frequency

# number of signal points
N = 300

# sample spacing
T = 1.0 / fs

# Generate the time vector properly
#t = np.arange(300) / fs
t = np.linspace(0.0, T*N, N)
#t = np.linspace(0.0, 5, N) #5 secs


signala = np.sin(2*np.pi*0.2*t) # with frequency of 0.2 Hertz
#plt.plot(t, signala, label='a')

signalb = np.sin(2*np.pi*1*t) # frequency 1Hz
plt.plot(t, signalb, label='b')

signalc = np.sin(2*np.pi*5*t) # frequency 5Hz
#plt.plot(t, signalc, label='c')

total = signala + signalb + signalc
plt.plot(t, total, label='total')

fc = np.array([0.75, 2.])  # Cut-off frequency of the filter
# 0.75 hz - 2 hz => 45bpm - 120bpm

w = fc / (fs / 2) # Normalize the frequency
b, a = signal.butter(5, w, 'bandpass')
output = signal.filtfilt(b, a, total)
plt.plot(t, output, label='filtered')
plt.legend()
plt.show()
```

```python
# Show fft
yf = fft(total)
xf = fftfreq(N, T)
#xf = fftshift(xf)
#yplot = fftshift(yf)

plt.plot(xf, 1.0/N * np.abs(yf))
plt.grid()
plt.show()
xf
```
